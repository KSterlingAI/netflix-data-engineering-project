Leer SILVER

```python
df = spark.table("silver.netflix_titles_clean")
```

KPI #1 Películas vs Series

```python
from pyspark.sql.functions import count

content_summary = df.groupBy("type") \
    .agg(count("*").alias("total_titles"))
```

```python
display(content_summary)
```

```python
content_summary.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("gold.content_summary")
```

KPI #2 Contenido por país

```python
country_summary = df.groupBy("country") \
    .agg(count("*").alias("total_titles")) \
    .orderBy("total_titles", ascending=False)
```

```python
display(country_summary)
```

```python
country_summary.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("gold.country_summary")
```

KPI #3 Contenido por año

```python
year_summary = df.groupBy("release_year") \
    .agg(count("*").alias("total_titles")) \
    .orderBy("release_year", ascending=False)
```

```python
display(year_summary)
```

```python
year_summary.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("gold.year_summary")
```
