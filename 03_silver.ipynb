Leer tabla BRONZE

```python
df = spark.table("bronze.netflix_titles")
```

```python
display(df)
```

ver SCHEMA

```python
df.printSchema()
```

Eliminar Duplicados

```python
clean_df = df.dropDuplicates()
```

Limpiar Nulos

```python
clean_df = clean_df.fillna({
    "director": "Unknown",
    "country": "Unknown",
    "cast": "Unknown",
    "rating": "Not Rated"
})
```

```python
display(clean_df)
```

Guardar Silver Layer

```python
clean_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("silver.netflix_titles_clean")
```

Validar SILVER Layer

```python
spark.sql("""
SELECT *
FROM silver.netflix_titles_clean
LIMIT 10
""").show()
```

Crear métricas rápidas

```python
from pyspark.sql.functions import count

clean_df.groupBy("type") \
    .agg(count("*").alias("total_titles")) \
    .show()
```
